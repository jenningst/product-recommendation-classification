```python
import joblib
import pandas as pd
```

```python
# import the fitted model, make predictions, and conduct error analysis on the wrong predictions
model = joblib.load('../models/logistic-regression-tuned.joblib')

train = pd.read_csv('../data/final/train.csv')
X = train.drop(labels=['w2v_feat'], axis=1)
reviews = train['review_text']

# make predictions
y_pred = model.predict(X)

error_df = pd.concat([X, pd.Series(y_pred, name='y_pred')], axis=1)
error_indexes = error_df[(error_df['target'] == 1) & (error_df['y_pred'] == 0)]
```
